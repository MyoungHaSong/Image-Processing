## 3.[픽셀 그룹 처리](#section3)
픽셀그룹처리란
출력 영상의 새로운 픽셀 값을 결정하기 위해 해당 픽셀 뿐아니라 그 주위의 이웃픽셀들도 함께 고려하는 공간영역 연산을 의미한다. Convoultion(Mask)라고 불리는 Filter를 이용하여 2차원 배열내의 가중치를 곱하고 그 값을 모두 더한 값으로 Convolution의 중앙에 있는 값을 대신한다. 

<img src="./images/day2/day2group.PNG" width = "500" height = "500">

컨볼루션은 2차원 배열로 다른이름으로 마스크(Mask), 필터(Filter), 템플릿(Template), 커널(Kernel)이라고 불리며,영상의 Smoothing, Sarpness, Edge Detection등등의 영상처리에 이용된다. 


픽셀그룹처리의 한가지 문제점.

가장자리(경계)를 처리하는데 어려움이 있다. 컨볼루션의 크기는 보통 1x1를 사용하지 않기 때문에 가장자리 부분을 처리할때 필터가 영상의 밖으로 나가기 때문에 나간부분에 대해서 연산을 진행할 수 없다. 이를 해결할 수 있는 방법은 크게 4가지가 있다.


<img src="./images/day2/day2padding.PNG" width = "800" height = "800">


1. 모든 가장자리를 0으로 처리를 한 뒤에 컨볼루션 연산을 진행한다. 
2. 경계값의 있는 픽셀 값을 한번 더 복사해준다.
3. 영상을 주기적인 신호로 해석하여 맞은편의 픽셀값을 복사
4. 가장자리 문제로 해결이 불가능한 부분을 제외하고 진행한 뒤 마지막에 input의 가장자리 부분을 기존영상의 값으로 복사

### Noise(잡음)
Noise(잡음)이란 영상내에서 원하지 않거나 흐름에 알맞지 않은 값으로 제거를 해야할 것들이다. 잡음은 크게 3가지 Impulse Noise, Gaussian Noise, Quantization Noise가 있다. 


- Impulse Noise는 이미지위에 소금과 후추를 뿌린 것처럼 보인다고 해서 Salt and Pepper Noise라고도 불린다. 0또는 255의 픽셀값과 같이 뚜렷하게 잘된 밝기를 갖는 Noise이다. 


<img src="./images/day2/day2impulse.PNG" width = "500" height = "500">



- Gaussian Noise는 가우시안 분포의 모양을 가지는 잡음으로 가장 일반적이고 자연적으로 랜덤하게 발생하는 잡음이다. 잡음이 가지는 값들의 분포가 Gaussian Distribution의 분포 모양을 띄고 있다. 

<img src="./images/day2/day2gaussian.PNG" width = "500" height = "500">


- Quantization Noise는 Quantization을 할때 원 아날로그 신호의 진폭값과 근사값 사이의 오차가 발생하고 복호화시에 잡음이나 왜곡같은 효과를 주어 양자화 잡음이라 하며, 품질저하의 요인이 된다.


<img src="./images/day2/day2noise.PNG" width = "700" height = "700">


첫번째 이미지는 지난시간에 살펴본 Sampling과 Quantization과정에서 나온 표이다. 이 표를 상에서 가로 t를 잘게 나누는 과정이 Sampling이고 그 신호를 세로로 일정한 양으로 측정하는 것이 Quantization이다.


두번째 이미지는 sampling한 신호를 quantization을 진행한 결과이다. 여기서는 0.1단위로 신호를 quantization을 진행 시킨 그림이다. 
이렇듯 Quantization시에 원 아날로그 신호의 진폭값과 근사값 사이의 오차가 발생하게 되고 복호화시에 잡음이나 왜곡같은 효과를 주어 양자화 잡이라고 한다. 세번째 이미지를 보면 해당 신호까지 quantization을 진행할때 검정색으로 빗금친부분만큼의 양자화를 진행하지 못했다. 이런 부분이 양자화 잡음으로 나타난다.

### Smoothing
 Smoothing은 위에서 살펴본 Noise를 제거하고 영상을 부드럽게 만들어주는 효과내는 기법을 말한다. 
   - Mean Filter(Smoothing)
        - MeanFilter는 Mask의 weight값이 모두 1을 가지고 weight의 개수만큼 나눠주어 평균값을 만드는 Filter를 의미한다. Mean filter 연산을 하고나면 이미지는 흐릿해지고 왜곽역시 흐려지지만 Noise를 없애는 효과와 영상을 매끄럽게 만들어주는 효과가 있다.
        
        
<img src="./images/day2/day2meanfilter.PNG" width = "500" height = "500">


   - Median Filter(Smoothing)
        - Median Filter는 Filter내의 값들을 순서대로 나열하고 그 중 중간에 있는 값을 대체해주는 Filter를 의미한다. Median Filter는 특히 Salt Peper Noise를 Clean하게 만드는데 아주 효과적인 Filter로 알려져있다.


<img src="./images/day2/day2median.PNG" width = "500" height = "500">


   - Gaussian Filter(Smoothing)
        - Gaussian Distribution은 모든 과학분야에서 나타나는 가장 보편적인 분포이다. Sample들의 값들이 평균이 Mu, 분산이 Sigma^2값을 갖는 분포로 아래 이미지와 같이 분포가 같다. 정규분포에서 평균값이 0으로 하여 유도한 분포이다. 


<img src="./images/day2/day2gaussianfilter.PNG" width = "500" height = "500">


Gaussian Filter는 위 사진과 같이 값이 있을때는 상관 없지만 평균점으로 부터 멀리있는 값이 나타날때 이 값에 Weight를 부여하여 평균근처의 값이 되게끔 만들어 Noise를 제거하는 Filter이다. 표준편차 Sigma값에 따라서 Smoothing정도가 달라지며 Sigma값이 높을수록 영상이 흐려진다.

# Image warping

<a id='section8'></a>

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import PiecewiseAffineTransform, warp,AffineTransform
from skimage import data


image = data.astronaut()
rows, cols = image.shape[0], image.shape[1]

src_cols = np.linspace(0, cols, 20)
src_rows = np.linspace(0, rows, 10)
src_rows, src_cols = np.meshgrid(src_rows, src_cols)
src = np.dstack([src_cols.flat, src_rows.flat])[0]

# add sinusoidal oscillation to row coordinates
dst_rows = src[:, 1] - np.sin(np.linspace(0, 3 * np.pi, src.shape[0])) * 50
dst_cols = src[:, 0]
dst_rows *= 1.5
dst_rows -= 1.5 * 50
dst = np.vstack([dst_cols, dst_rows]).T


tform = PiecewiseAffineTransform()
tform.estimate(src, dst)

out_rows = image.shape[0] - 1.5 * 50
out_cols = cols
out = warp(image, tform, output_shape=(out_rows, out_cols))

fig, ax = plt.subplots()
ax.imshow(out)
ax.plot(tform.inverse(src)[:, 0], tform.inverse(src)[:, 1], '.b')
ax.axis((0, out_cols, out_rows, 0))
plt.show()
```

# High Pass Filter

<a id='section10'></a>

```python
from skimage.io import imread
import matplotlib.pyplot as plt
import scipy.fftpack as fp

im = np.mean(data.chelsea(), axis=2) # assuming an RGB image

F1 = fp.fft2((im).astype(float))
F2 = fp.fftshift(F1)
(w, h) = im.shape
half_w, half_h = int(w/2), int(h/2)

# high pass filter
n = 25
F2[half_w-n:half_w+n+1,half_h-n:half_h+n+1] = 0 # select all but the first 50x50 (low) frequencies

im1 = fp.ifft2(fp.ifftshift(F2)).real
fig, axes = plt.subplots(1,2,figsize = (10,10))
ax = axes.ravel()

ax[0].imshow(im, cmap= plt.cm.gray)
ax[1].imshow(im1)

for a in ax:
    a.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.morphology import disk
from skimage.filters import rank


image = data.coins()
selem = disk(20)

percentile_result = rank.mean_percentile(image, selem=selem, p0=.1, p1=.9)
bilateral_result = rank.mean_bilateral(image, selem=selem, s0=500, s1=500)
normal_result = rank.mean(image, selem=selem)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10),
                         sharex=True, sharey=True)
ax = axes.ravel()

titles = ['Original', 'Percentile mean', 'Bilateral mean', 'Local mean']
imgs = [image, percentile_result, bilateral_result, normal_result]
for n in range(0, len(imgs)):
    ax[n].imshow(imgs[n], cmap=plt.cm.gray)
    ax[n].set_title(titles[n])
    ax[n].axis('off')

plt.tight_layout()
plt.show()
```
