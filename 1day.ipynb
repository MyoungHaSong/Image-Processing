## 1. [영상처리 기초지식](#section1)


**영상**이란 넓은 의미에서 시각적 정보를 기록하고 표현하는 하나의 방식인 그림이나 사진을 뜻한다.
영상은 2차원 함수 $f(x,y)$ 또는 배열로 표현할 수 있으며, 여기서 $x$와 $y$는 공간의 좌표이다. 영상처리 알고리즘들을 분류하는 기본적인 분류방법은 4가지로 **포인트 처리(Point Opterator)**, **영역처리(Regional Operator)**, **기하학적 처리(Geometric Operator)**, **프레임처리(Frame Operator)** 가 있다.



---

### 1-1. Image Processing
- 포인트처리(point operator):
    - 포인트 처리란 2차원 배열에서 한 점(element)에 대해서 처리를 하는 기술을 말한다. 한점에 대한 연산을 모든 화소값에 대해서 처리를 하는것이 포인트 처리이다. 


- 영역처리(reginal operator)
    - 영역 처리란 filter(Mask,Convolution)를 이용하여 마스크의 중앙에 있는 값을 주변 pixel의 값을 이용하여 함께 처리하는 기술이다. 


- 기하학적 처리(geometric operator)
    - 영상에 대해 회전, 왜곡, 평행이동등의 기하학적 변환을 가하는 기술을 말한다. 


- 프레임처리(frame operator)
    - 두 개 이상의 서로 다른 영상을 포함한 영상간의 연산을 바탕으로 새로운 화소 값을 처리하는 기술

---

### 1-2. 화소(pixel)이란
- 2차원 배열에서 한개의 점에 해당하는 값 보통 0~255 256단계의 값을 가진다

<img src="./images/day1/day1pixel_image.PNG">

---   

### 1-3. 색의 표현
- RGB(Red, Green, Blue)의 색상의 혼합으로 모든 색을 표현한다. 각축(RGB)은 0~255까지의 값을 가진다.
- HIS(Hue, Intensity, Saturation) 색상, 명암, 채도의 공간으로 각축은 0~255까지의 값을 가진다.

<img src="./images/day1/day1colorimage.PNG">

---

        
### 1-4. 양자화와 디지털화(표본화)
- 아날로그 이미지는 컴퓨터에서 직접 처리하거나 저장할 수 없기 때문에 표본화(Sampling) 및 양자화(Quantization) 과정을 거쳐 이미지로 변환된다.
    
    - 표본화 : 아날로그 이미지의 연속적인 위치 데이터를 불연속적인 위치 데이터로 변환하는 과정
    - 양자화 : 
        - 연속적인 색상/밝기 데이터를 불연속적인 디지털 데이터로 변환하는 과정
        - 각 화소의 밝기 또는 색상을 컴퓨터에서 인지할 수 있는 숫자로 표현하는 과정
        - Ex) 픽셀의 밝기 값을 8bit 256단계로 제한한다면 밝기를 0에서 255까지의 숫자로 명암을 표현
        - Ex) X선 이미지의 경우 1025레벨(10bit)

## 2. [히스토그램](#section2)
히스토그램은 어떤 영상의 명암값의 정보, 특성을 보여주기 위해 사용되는 중요한 도구이다. 히스토그램을 사용하여 영상의 명암대비 및 명암 값의 분포등의 영상의 구성을 자세하게 알 수 있다. 히스토그램에서 X축은 화소가 가질 수 있는 명암의 값, y축에는 빈도로 표현한다. 

---

### 2-1. Thresholding
- 영상분석에서 자주 사용되는것이 이진화라는 것이다.이진화는 픽셀값을 0또는 1로 바꾸는 과정을 말하는데, 이진화를 이용하여서 배경과 물체를 분리하거나 일정한 밝기 값 이상의 픽셀들만 추출, Image Segmentation 등 영상의 전처리 단계에서 사용되는 기법이다. 가장 쉽게 적용할 수 있는 기법이 히스토그램을 이용하여 분포를 확인하고 임계값(Thresholding)을 설정하여 그 임계값을 기준으로 0또는 1로 나누는 방법이다. 

<img src="./images/day1/day1histogram.PNG">


히스토그램을 보고 봉우리에서 최하점을 보고 선택할수도 있겠지만 봉우리가 뚜렷하지 않은 경우엔 분포를 보고 파악하기 힘들 것이다. 그리고 수많은 데이터를 다뤄야 하는데 일일이 히스토그램을 보고 데이터를 이진화 처리 할 수 없기 때문에 알고리즘을 이용하여서 이진화를 진행해야 한다.

### 2-2. Histogram Equalization
히스토그램을 이용해서 영상의 화질을 개선하는 것이 가능하다. 그 기법 이름을 히스토그램의 균일화(Histogram Equalization)이다. 


1) cdf 누적합을 구한다.


2) 수식에 대입한다.

위에 그림에서 볼 수 있듯이 히스토그램 균일화를 하게 되면 기존에 표현하지 못한 부분까지 조금 더 잘 표현할 수 있다. 하지만 이런 히스토그램 균일화의 단점이 있는데, 바로 전체 픽셀에 대해서 진행하기 때문에 명암값의 차이가 없는 부분도 차이가 있는 것 처럼 보이거나 또는 명암값의 차이가

이런 단점을 극복하게 나온 것이 바로 Adaptive Histogram Equalization이다. Adaptive Histogram Equalization은 전체 픽셀 명암값을 대상으로 하는 것이 아니라, 영상 전체를 분할을 하고 분할한 영상에 대해서 Equaliation을 하여 부분적으로 균일화를 진행하게 된다.


부분적으로 진행하게 될 경우, 기존에 큰 차이가 없는 지역에 대해서도 적절하게 잘 표현하면서도 기존영상보다 훨씬 화질이 개선된 영상이 나타나는 것을 볼 수 있다.

## 3. [픽셀 점 처리](#section3)

# Histogram Equalization

<a id='section6'></a>

```python
matplotlib.rcParams['font.size'] = 8

def plot_img_and_hist(image, axes, bins=256):
    """Plot an image along with its histogram and cumulative histogram.

    """
    image = img_as_float(image)
    ax_img, ax_hist = axes
    ax_cdf = ax_hist.twinx()

    # Display image
    ax_img.imshow(image, cmap=plt.cm.gray)
    ax_img.set_axis_off()

    # Display histogram
    ax_hist.hist(image.ravel(), bins=bins, histtype='step', color='black')
    ax_hist.ticklabel_format(axis='y', style='scientific', scilimits=(0, 0))
    ax_hist.set_xlabel('Pixel intensity')
    ax_hist.set_xlim(0, 1)
    ax_hist.set_yticks([])

    # Display cumulative distribution
    img_cdf, bins = exposure.cumulative_distribution(image, bins)
    ax_cdf.plot(bins, img_cdf, 'r')
    ax_cdf.set_yticks([])

    return ax_img, ax_hist, ax_cdf


# Load an example image
img = data.moon()

# Contrast stretching
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

# Equalization
img_eq = exposure.equalize_hist(img)

# Adaptive Equalization
img_adapteq = exposure.equalize_adapthist(img, clip_limit=0.03)

# Display results
fig = plt.figure(figsize=(8, 5))
axes = np.zeros((2, 4), dtype=np.object)
axes[0, 0] = fig.add_subplot(2, 4, 1)
for i in range(1, 4):
    axes[0, i] = fig.add_subplot(2, 4, 1+i, sharex=axes[0,0], sharey=axes[0,0])
for i in range(0, 4):
    axes[1, i] = fig.add_subplot(2, 4, 5+i)

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img, axes[:, 0])
ax_img.set_title('Low contrast image')

y_min, y_max = ax_hist.get_ylim()
ax_hist.set_ylabel('Number of pixels')
ax_hist.set_yticks(np.linspace(0, y_max, 5))

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_rescale, axes[:, 1])
ax_img.set_title('Contrast stretching')

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_eq, axes[:, 2])
ax_img.set_title('Histogram equalization')

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_adapteq, axes[:, 3])
ax_img.set_title('Adaptive equalization')

ax_cdf.set_ylabel('Fraction of total intensity')
ax_cdf.set_yticks(np.linspace(0, 1, 5))

# prevent overlap of y-axis labels
fig.tight_layout()
plt.show()
```

# 이진화

<a id='section10'></a>

```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray

image = rgb2gray(np.array(Image.open("./lion.jpg")))
thresh = threshold_otsu(image)
binary = image > thresh

# otsu알고리즘을 이용하여 다른 이미지에 thresholding
fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

## image slicing

<a id='section11'></a>

```python
image = rgb2gray(np.array(Image.open("./lion.jpg")))
img1 = np.copy(image)

# 특정 구역을 정하여 그부분을 1로 만들어 준다.

for i in range(512):
    for j in range(512):
        if (img1[i,j] >= 0.2 and img1[i,j] <0.6) or (img1[i,j] >= 0.8 and img1[i,j] <=1) :
            img1[i,j] = 1
fig , ax = plt.subplots(nrows = 1 , ncols = 2 , figsize = (8,5))

ax[0].imshow(image, cmap = plt.cm.gray)
ax[0].axis('off')
ax[0].set_title("Original image")
ax[1].imshow(img1, cmap = plt.cm.gray)
ax[1].axis('off')
ax[1].set_title("Contrast Slicing")

plt.show()
```

# gamma correlation

<a id='section12'></a>

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure

matplotlib.rcParams['font.size'] = 8


def plot_img_and_hist(image, axes, bins=256):
    """Plot an image along with its histogram and cumulative histogram.

    """
    image = img_as_float(image)
    ax_img, ax_hist = axes
    ax_cdf = ax_hist.twinx()

    # Display image
    ax_img.imshow(image, cmap=plt.cm.gray)
    ax_img.set_axis_off()

    # Display histogram
    ax_hist.hist(image.ravel(), bins=bins, histtype='step', color='black')
    ax_hist.ticklabel_format(axis='y', style='scientific', scilimits=(0, 0))
    ax_hist.set_xlabel('Pixel intensity')
    ax_hist.set_xlim(0, 1)
    ax_hist.set_yticks([])

    # Display cumulative distribution
    img_cdf, bins = exposure.cumulative_distribution(image, bins)
    ax_cdf.plot(bins, img_cdf, 'r')
    ax_cdf.set_yticks([])

    return ax_img, ax_hist, ax_cdf


# Load an example image
img = data.moon()

# low-Gamma-image
low_gamma_image = exposure.adjust_gamma(img, 0.5)

# high-gamma-image
high_gamma_image = exposure.adjust_gamma(img, 2)

# Display results
fig = plt.figure(figsize=(8, 5))
axes = np.zeros((2, 3), dtype=np.object)
axes[0, 0] = plt.subplot(2, 3, 1)
axes[0, 1] = plt.subplot(2, 3, 2, sharex=axes[0, 0], sharey=axes[0, 0])
axes[0, 2] = plt.subplot(2, 3, 3, sharex=axes[0, 0], sharey=axes[0, 0])
axes[1, 0] = plt.subplot(2, 3, 4)
axes[1, 1] = plt.subplot(2, 3, 5)
axes[1, 2] = plt.subplot(2, 3, 6)

ax_img, ax_hist, ax_cdf = plot_img_and_hist(low_gamma_image, axes[:, 0])
ax_img.set_title('Low gamma image')

y_min, y_max = ax_hist.get_ylim()
ax_hist.set_ylabel('Number of pixels')
ax_hist.set_yticks(np.linspace(0, y_max, 5))

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img, axes[:, 1])
ax_img.set_title('Original image')

ax_img, ax_hist, ax_cdf = plot_img_and_hist(high_gamma_image, axes[:, 2])
ax_img.set_title('High gamma image')

ax_cdf.set_ylabel('Fraction of total intensity')
ax_cdf.set_yticks(np.linspace(0, 1, 5))

# prevent overlap of y-axis labels
fig.tight_layout()
plt.show()
```
